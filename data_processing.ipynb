```python
import pandas as pd
```

```python
df = pd.read_csv("merged.csv")
```

```python
df
```

```python
mask = (df['price']!='price')
df = df[mask]
df
```

```python
df.size
```

```python
finalDF = df.copy()
finalDF
```

```python
finalDF
```

```python
month_year_split = df['year'].str.split("\n", n=1, expand=True)
newMask = (month_year_split[0]!='year')
month_year_split = month_year_split.loc[:, mask]
month_year_split
```

```python
finalDF['year'] = month_year_split[1]
```

```python
finalDF['month'] = month_year_split[0]
```

```python
finalDF
```

```python
len(finalDF)
```

```python
origin_encode = pd.get_dummies(finalDF.origin, dummy_na=True, prefix="Origin")
destination_encode = pd.get_dummies(finalDF.destination, dummy_na=True, prefix="Destination")
month_encode = pd.get_dummies(finalDF.month, dummy_na=True, prefix="Month")
```

```python
# slc = (df['origin']=='SLC ')
# finalDF[slc]
```

```python
one_hot_df = finalDF.copy()
one_hot_df = pd.concat([one_hot_df, origin_encode], axis=1)
one_hot_df = pd.concat([one_hot_df, destination_encode], axis=1)
one_hot_df = pd.concat([one_hot_df, month_encode], axis=1)

one_hot_df.head(5)
```

```python
one_hot_df.columns
```

```python
one_hot_df.dtypes
```

```python
cols_to_drop = ['origin', 'destination', 'month']
one_hot_df = one_hot_df.drop(cols_to_drop, axis=1)
```

```python
one_hot_df.to_pickle("./finalDF.pkl")
```
